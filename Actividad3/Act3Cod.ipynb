```python
#Se importan las librerías.
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
sentinels = {'Precipitación': ['Nulo'],'Evaporación':['Nulo'],'TemperaturaMáxima':['Nulo'],'TemperaturaMínima':['Nulo']}
```

```python
#Se lee la tabla.
df0 = pd.read_csv('Alamos.txt', skiprows = 20, sep='\s+', header = None, names = ["Día/Mes/Año", "Precipitación", "Evaporación", "TemperaturaMáxima", "TemperaturaMínima"], na_values=sentinels)
```

```python
#Se cambia las fechas a variable temporal.
df0['Fecha'] = pd.to_datetime(df0.apply(lambda x: x['Día/Mes/Año'], 1),dayfirst =True)
```

```python
#Se elimina la primera columna.
df0 = df0.drop(['Día/Mes/Año'],axis = 1)
```

```python
#Se crearán columnas para los meses y años.
df0['Mes'] = df0['Fecha'].dt.month
df0['Año'] = df0['Fecha'].dt.year
df0.head()
```

```python
#Se le asigna un número a cada mes.
df0_Enero = df0[df0.Mes==1]
df0_Febrero = df0[df0.Mes==2]
df0_Marzo = df0[df0.Mes==3]
df0_Abril = df0[df0.Mes==4]
df0_Mayo = df0[df0.Mes==5]
df0_Junio = df0[df0.Mes==6]
df0_Julio = df0[df0.Mes==7]
df0_Agosto = df0[df0.Mes==8]
df0_Septiembre = df0[df0.Mes==9]
df0_Octubre = df0[df0.Mes==10]
df0_Noviembre = df0[df0.Mes==11]
df0_Diciembre = df0[df0.Mes==12]
```

```python
#Se hace la sumatoria de las precipitaciones.
#Ps es el promedio de precipitación de cada mes.
Ps_Enero = df0_Enero.Precipitación.sum()
print("Enero = ",Ps_Enero)
Ps_Febrero = df0_Febrero.Precipitación.sum()
print("Febrero = ",Ps_Febrero)
Ps_Marzo = df0_Marzo.Precipitación.sum()
print("Marzo = ",Ps_Marzo)
Ps_Abril = df0_Abril.Precipitación.sum()
print("Abril = ",Ps_Abril)
Ps_Mayo = df0_Mayo.Precipitación.sum()
print("Mayo = ",Ps_Mayo)
Ps_Junio = df0_Junio.Precipitación.sum()
print("Junio = ",Ps_Junio)
Ps_Julio = df0_Julio.Precipitación.sum()
print("Julio = ",Ps_Julio)
Ps_Agosto = df0_Agosto.Precipitación.sum()
print("Agosto = ",Ps_Agosto)
Ps_Septiembre = df0_Septiembre.Precipitación.sum()
print("Septiembre = ",Ps_Septiembre)
Ps_Octubre = df0_Octubre.Precipitación.sum()
print("Octubre = ",Ps_Octubre)
Ps_Noviembre = df0_Noviembre.Precipitación.sum()
print("Noviembre = ",Ps_Noviembre)
Ps_Diciembre = df0_Diciembre.Precipitación.sum()
print("Diciembre = ",Ps_Diciembre)
```

```python
#Se grafica el promedio de precipitación de cada mes:
DesMois = ('Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre')
#Para que la tabla abarque los valores de los promedio de las precipitaciones.
y_pos = np.arange(len(DesMois))
PreTot = [Ps_Enero,Ps_Febrero,Ps_Marzo,Ps_Abril,Ps_Mayo,Ps_Junio,Ps_Julio,Ps_Agosto,Ps_Septiembre,Ps_Octubre,Ps_Noviembre,Ps_Diciembre]

#Para hacer la gráfica:
plt.bar(y_pos, PreTot,align='center', alpha=0.3)

#Poner los meses en el eje de las "x".
plt.xticks(y_pos, DesMois,rotation=45)
plt.ylabel('Promedio de precipitación por mes (mm)')
plt.title('Promedio de precipitación por meses (1961-1984)')
plt.savefig('RainMeses',plt=2000)
plt.show()
```

```python
#Se puede observar que el mes con mayor cantidad de precipitaciones es Agosto.
```

```python
#Con el objetivo de calcular cuáles han sido los años más húmedos, ésto se ha basado en la suma de precipitaciones por año
#Se creó un arreglo con los años y con la precipitación total por año
#Begin es el año donde empezamos a contar, en un rango de 23 años hasta el 1984.
#Año es el arreglo con los 23 años.
#Year es el arreglo con la suma de precipitaciones por año de los 23 años.
Begin = 1961
Year = [Begin + i for i in range(0, 24)]
Years = [ df0[df0.Año==(Begin + i)].Precipitación.sum() for i in range(0, 24)]
```

```python
#Se crea un nuevo Data Frame para los años y las precipitaciones totales.
d0 = {'Año': Year, 'Precipitación total': Years}
df1 = pd.DataFrame(data=d0)
df1.head()
```

```python
df1.tail()
```

```python
#Graficamos las precipitaciones totales por año.
fecha = Year
T = np.arange(len(fecha))
Lluvia = Years

plt.bar(T, Lluvia, width=0.4, align='center', alpha=0.4)
plt.xticks(T, fecha, size='xx-small', rotation=90)
plt.ylabel('Precipitación (mm)')
plt.title('Precipitación total por años (1961-1984)')
plt.savefig('PreTotalAños',plt=2000)
plt.show()
```

```python
#Se concluye que el año más húmedo (es decir, con más lluvias) es 1981: con 1155 mm, mientras que el más seco (el de menos lluvias) fue 1979: con 330 mm.
```

```python
#Se seleccionan las temperaturas máximas y mínimas de los meses de invierno (diciembre, enero y febrero)
TempInv = pd.DataFrame(df0.loc[df0['Mes'].isin([12,1,2])])
TempInv.head()
```

```python
#Se crean arreglos para la temperatura promedio máxima y mínima de cada año
#InvMax 
InvMax = [TempInv[TempInv.Año==(Begin + i)].TemperaturaMáxima.mean() for i in range(0,24)]
InvMin = [TempInv[TempInv.Año==(Begin + i)].TemperaturaMínima.mean() for i in range(0,24)]
```

```python
#Creamos otro DataFrame para las temperaturas promedio por año
d3 = {'Año': Year, 'TemperaturaMáximaInvierno': InvMax, 'TemperaturaMínimaInvierno': InvMin }
df3 = pd.DataFrame(data=d3)
df3
```

```python
#Se crean gráficas para la temperatura promedio máxima y mínima en invierno por cada año.
N = np.arange(len(fecha))

plt.bar(N,InvMax,width=0.4,align='center',alpha=0.7,color='C')
plt.xticks(N,Year,size='small',color='c',rotation=90)

plt.bar(N,InvMin,width=0.4,align='center',alpha=0.6,color='y')
plt.xticks(N,Year,size='small',color='k',rotation=90)

plt.ylabel('Temperatura (°C)')
plt.title('Temperatura promedio máxima y mínima en invierno (1961-1984)')
plt.gca().legend(('Temperatura máxima','Temperatura mínima'),frameon=False,loc='upper left')
plt.savefig('Temperaturasprom_invierno',plt=2000)
plt.show()
```

```python
#Se concluye que la temperatura mínima fue en 1979 y la temperatura máxima fue en 1977.
```
